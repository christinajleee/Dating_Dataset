```python
# Summary Statistics for Dating App Dataset
import pandas as pd
import numpy as np

print("DATING APP DATASET - SUMMARY STATISTICS")
print("=" * 80)

# Load data (assuming df is already loaded)
# df = pd.read_csv('dating_app_behavior_dataset_extended1.csv')
# df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# --------------------------------------
# 1. NUMERIC VARIABLES SUMMARY
# --------------------------------------
print("\nðŸ”¢ NUMERIC VARIABLES SUMMARY")
print("-" * 60)

# Get numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Create comprehensive summary statistics
summary_stats = df[numeric_cols].describe().T

# Add additional statistics
summary_stats['median'] = df[numeric_cols].median()
summary_stats['mode'] = df[numeric_cols].mode().iloc[0] if len(df[numeric_cols].mode()) > 0 else np.nan
summary_stats['range'] = summary_stats['max'] - summary_stats['min']
summary_stats['variance'] = df[numeric_cols].var()
summary_stats['skewness'] = df[numeric_cols].skew()
summary_stats['missing_count'] = df[numeric_cols].isnull().sum()
summary_stats['missing_pct'] = (df[numeric_cols].isnull().sum() / len(df) * 100).round(2)

# Reorder columns for better readability
column_order = ['count', 'mean', 'median', 'mode', 'std', 'min', 'max', 'range', 
                '25%', '50%', '75%', 'variance', 'skewness', 'missing_count', 'missing_pct']
summary_stats = summary_stats[column_order]

# Round numeric values for readability
summary_stats = summary_stats.round(2)

print(summary_stats)

# --------------------------------------
# 2. CATEGORICAL VARIABLES SUMMARY
# --------------------------------------
print("\n\n CATEGORICAL VARIABLES SUMMARY")
print("-" * 60)

# Get categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

if categorical_cols:
    cat_summary = []
    
    for col in categorical_cols:
        col_data = df[col]
        unique_values = col_data.nunique()
        most_common = col_data.mode().iloc[0] if len(col_data.mode()) > 0 else "N/A"
        most_common_count = col_data.value_counts().iloc[0] if len(col_data.value_counts()) > 0 else 0
        most_common_pct = (most_common_count / len(col_data) * 100) if len(col_data) > 0 else 0
        missing_count = col_data.isnull().sum()
        missing_pct = (missing_count / len(col_data) * 100)
        
        cat_summary.append({
            'Variable': col,
            'Unique_Values': unique_values,
            'Most_Common': most_common,
            'Most_Common_Count': most_common_count,
            'Most_Common_Pct': round(most_common_pct, 2),
            'Missing_Count': missing_count,
            'Missing_Pct': round(missing_pct, 2)
        })
    
    cat_summary_df = pd.DataFrame(cat_summary)
    print(cat_summary_df.to_string(index=False))
else:
    print("No categorical variables found.")

# --------------------------------------
# 3. TOP-LEVEL DATASET OVERVIEW
# --------------------------------------
print("\n\n DATASET OVERVIEW")
print("-" * 60)

overview_stats = {
    'Total Records': len(df),
    'Total Variables': len(df.columns),
    'Numeric Variables': len(numeric_cols),
    'Categorical Variables': len(categorical_cols),
    'Total Missing Values': df.isnull().sum().sum(),
    'Missing Value %': round((df.isnull().sum().sum() / (len(df) * len(df.columns))) * 100, 2),
    'Complete Records': len(df.dropna()),
    'Complete Records %': round((len(df.dropna()) / len(df)) * 100, 2)
}

for key, value in overview_stats.items():
    print(f"{key:25}: {value}")

# --------------------------------------
# 4. KEY VARIABLE DISTRIBUTIONS
# --------------------------------------
print("\n\n KEY DATING VARIABLES - DETAILED BREAKDOWN")
print("-" * 60)

# Focus on key dating-related variables
key_vars = ['age', 'height_cm', 'weight_kg', 'mutual_matches', 'likes_received', 
            'swipe_right_ratio', 'app_usage_time_min', 'bio_length', 'profile_pics_count']

key_available = [var for var in key_vars if var in df.columns]

if key_available:
    key_stats = df[key_available].describe().T
    key_stats['median'] = df[key_available].median()
    key_stats['missing_pct'] = (df[key_available].isnull().sum() / len(df) * 100).round(2)
    
    # Reorder and round
    key_order = ['count', 'mean', 'median', 'std', 'min', '25%', '50%', '75%', 'max', 'missing_pct']
    key_stats = key_stats[key_order].round(2)
    
    print("Key Variables Summary:")
    print(key_stats)

# --------------------------------------
# 5. SUCCESS METRICS ANALYSIS
# --------------------------------------
print("\n\n SUCCESS METRICS ANALYSIS")
print("-" * 60)

success_vars = ['mutual_matches', 'likes_received', 'swipe_right_ratio']
success_available = [var for var in success_vars if var in df.columns]

if success_available:
    print("Success Variable Percentiles:")
    percentiles = [10, 25, 50, 75, 90, 95, 99]
    success_percentiles = df[success_available].quantile([p/100 for p in percentiles]).T
    success_percentiles.columns = [f'P{p}' for p in percentiles]
    success_percentiles = success_percentiles.round(2)
    print(success_percentiles)

# --------------------------------------
# 6. CORRELATION MATRIX (TOP CORRELATIONS)
# --------------------------------------
print("\n\n TOP CORRELATIONS")
print("-" * 60)

if len(numeric_cols) > 1:
    corr_matrix = df[numeric_cols].corr()
    
    # Get top positive and negative correlations (excluding self-correlations)
    corr_pairs = []
    for i in range(len(corr_matrix.columns)):
        for j in range(i+1, len(corr_matrix.columns)):
            var1 = corr_matrix.columns[i]
            var2 = corr_matrix.columns[j]
            correlation = corr_matrix.iloc[i, j]
            if not np.isnan(correlation):
                corr_pairs.append({
                    'Variable_1': var1,
                    'Variable_2': var2,
                    'Correlation': round(correlation, 3)
                })
    
    # Sort by absolute correlation
    corr_df = pd.DataFrame(corr_pairs)
    corr_df['Abs_Correlation'] = corr_df['Correlation'].abs()
    corr_df = corr_df.sort_values('Abs_Correlation', ascending=False)
    
    print("Top 10 Strongest Correlations:")
    print(corr_df.head(10)[['Variable_1', 'Variable_2', 'Correlation']].to_string(index=False))

# --------------------------------------
# 7. DEMOGRAPHIC BREAKDOWNS
# --------------------------------------
print("\n\n DEMOGRAPHIC ANALYSIS ")
print("-" * 60)

demo_vars = ['gender', 'sexual_orientation', 'education_level', 'income_bracket', 'body_type']
demo_available = [var for var in demo_vars if var in df.columns]

for var in demo_available[:3]:  # Show first 3 to avoid clutter
    print(f"\n{var.upper()} Distribution:")
    value_counts = df[var].value_counts()
    percentages = (df[var].value_counts(normalize=True) * 100).round(1)
    
    for category in value_counts.index[:5]:  # Top 5 categories
        count = value_counts[category]
        pct = percentages[category]
        print(f"  {category:20}: {count:6,} ({pct:5.1f}%)")
    
    if len(value_counts) > 5:
        remaining = len(value_counts) - 5
        print(f"  ... and {remaining} other categories")

print("\n" + "=" * 80)
print("ðŸ“Š Summary Statistics Complete!")
print(f"Dataset analyzed: {len(df):,} records Ã— {len(df.columns)} variables")
```

```python
# Import packages
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

try:
    # Load the extended dataset
    df = pd.read_csv('dating_app_behavior_dataset_extended1.csv')
    print(f"Dataset loaded successfully. Shape: {df.shape}")
except FileNotFoundError:
    print("Error: Dataset file not found. Please check the file path.")
    exit()

# Optional: clean column names (if needed)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Display basic info about the dataset
print("\nDataset columns:", df.columns.tolist())
print("\nFirst few rows:")
print(df.head())

# Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# --------------------------------------
# 1. App Usage Over Time (Last Active Hour)
# --------------------------------------
if 'last_active_hour' in df.columns:
    plt.figure(figsize=(10, 5))
    # Handle potential NaN values and convert to numpy array for compatibility
    last_active_clean = df['last_active_hour'].dropna()
    if len(last_active_clean) > 0:
        try:
            # Convert to numpy array to avoid pandas indexing issues
            hour_data = last_active_clean.values
            sns.histplot(hour_data, bins=24, kde=True)
            plt.title('User Activity by Hour of Day')
            plt.xlabel('Hour of Day')
            plt.ylabel('User Count')
            plt.xlim(0, 23)  # Ensure proper hour range
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Error with seaborn histplot: {e}")
            # Fallback to matplotlib
            plt.hist(hour_data, bins=24, alpha=0.7, edgecolor='black')
            plt.title('User Activity by Hour of Day')
            plt.xlabel('Hour of Day')
            plt.ylabel('User Count')
            plt.xlim(0, 23)
            plt.grid(True, alpha=0.3)
            plt.tight_layout()
            plt.show()
    else:
        print("No valid data for last_active_hour")
else:
    print("Column 'last_active_hour' not found in dataset")

# --------------------------------------
# 2. Income Bracket vs Education Level
# --------------------------------------
if 'income_bracket' in df.columns and 'education_level' in df.columns:
    plt.figure(figsize=(12, 6))
    # Remove NaN values for cleaner plotting
    df_clean = df.dropna(subset=['income_bracket', 'education_level']).copy()
    if len(df_clean) > 0:
        try:
            # Reset index to avoid indexing issues
            df_clean = df_clean.reset_index(drop=True)
            sns.countplot(data=df_clean, x='income_bracket', hue='education_level')
            plt.xticks(rotation=45, ha='right')  # Better rotation alignment
            plt.title('Income Bracket vs Education Level')
            plt.xlabel('Income Bracket')
            plt.ylabel('Count')
            plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Error with seaborn countplot: {e}")
            # Fallback to pandas crosstab with matplotlib
            crosstab = pd.crosstab(df_clean['income_bracket'], df_clean['education_level'])
            crosstab.plot(kind='bar', figsize=(12, 6))
            plt.title('Income Bracket vs Education Level')
            plt.xlabel('Income Bracket')
            plt.ylabel('Count')
            plt.legend(title='Education Level', bbox_to_anchor=(1.05, 1), loc='upper left')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
    else:
        print("No valid data for income_bracket and education_level combination")
else:
    print("Required columns for income/education analysis not found")

# --------------------------------------
# 3. Match Outcome Distribution
# --------------------------------------
if 'match_outcome' in df.columns:
    plt.figure(figsize=(8, 5))
    # Handle NaN values
    match_outcome_clean = df['match_outcome'].dropna()
    if len(match_outcome_clean) > 0:
        try:
            # Create a temporary dataframe for seaborn
            temp_df = pd.DataFrame({'match_outcome': match_outcome_clean}).reset_index(drop=True)
            sns.countplot(data=temp_df, x='match_outcome')
            plt.title('Distribution of Match Outcomes')
            plt.xlabel('Match Outcome')
            plt.ylabel('Number of Users')
            plt.xticks(rotation=45, ha='right')  # In case outcome names are long
            plt.tight_layout()
            plt.show()
        except Exception as e:
            print(f"Error with seaborn countplot: {e}")
            # Fallback to matplotlib
            outcome_counts = match_outcome_clean.value_counts()
            plt.bar(outcome_counts.index, outcome_counts.values)
            plt.title('Distribution of Match Outcomes')
            plt.xlabel('Match Outcome')
            plt.ylabel('Number of Users')
            plt.xticks(rotation=45, ha='right')
            plt.tight_layout()
            plt.show()
        
        # Correlation with match outcome (numeric columns only)
        numeric_cols = df.select_dtypes(include=[np.number])
        if 'match_outcome' in numeric_cols.columns:
            correlation = numeric_cols.corr()['match_outcome'].sort_values(ascending=False)
            print("\nCorrelation with Match Outcome:")
            print(correlation)
        else:
            print("\nMatch outcome is not numeric - cannot calculate correlation")
            print("Unique match outcomes:", df['match_outcome'].unique())
    else:
        print("No valid data for match_outcome")
else:
    print("Column 'match_outcome' not found in dataset")

# --------------------------------------
# 4. Swipe/Like/Match Behavior by Gender & Orientation
# --------------------------------------
required_cols = ['gender', 'swipe_right_ratio', 'sexual_orientation', 'likes_received', 'mutual_matches']
missing_cols = [col for col in required_cols if col not in df.columns]

if not missing_cols:
    # Check if we have enough data
    df_behavior = df.dropna(subset=required_cols).copy().reset_index(drop=True)
    if len(df_behavior) > 0:
        try:
            fig, axes = plt.subplots(1, 3, figsize=(18, 5))
            
            # Plot 1: Swipe Right Ratio
            sns.boxplot(data=df_behavior, x='gender', y='swipe_right_ratio', 
                       hue='sexual_orientation', ax=axes[0])
            axes[0].set_title('Swipe Right Ratio by Gender & Sexual Orientation')
            axes[0].set_ylabel('Swipe Right Ratio')
            
            # Plot 2: Likes Received
            sns.boxplot(data=df_behavior, x='gender', y='likes_received', 
                       hue='sexual_orientation', ax=axes[1])
            axes[1].set_title('Likes Received by Gender & Sexual Orientation')
            axes[1].set_ylabel('Likes Received')
            
            # Plot 3: Mutual Matches
            sns.boxplot(data=df_behavior, x='gender', y='mutual_matches', 
                       hue='sexual_orientation', ax=axes[2])
            axes[2].set_title('Mutual Matches by Gender & Sexual Orientation')
            axes[2].set_ylabel('Mutual Matches')
            
            # Adjust legends and layout
            for i, ax in enumerate(axes):
                ax.set_xlabel('Gender')
                # Only show legend on the last plot to avoid clutter
                if i < 2:
                    if ax.get_legend():
                        ax.get_legend().remove()
                else:
                    if ax.get_legend():
                        ax.legend(title='Sexual Orientation', bbox_to_anchor=(1.05, 1), loc='upper left')
            
            plt.tight_layout()
            plt.show()
            
        except Exception as e:
            print(f"Error with seaborn boxplots: {e}")
            print("Using matplotlib fallback for behavior analysis...")
            
            # Fallback to grouped bar charts
            fig, axes = plt.subplots(1, 3, figsize=(18, 5))
            
            metrics = ['swipe_right_ratio', 'likes_received', 'mutual_matches']
            titles = ['Swipe Right Ratio', 'Likes Received', 'Mutual Matches']
            
            for i, (metric, title) in enumerate(zip(metrics, titles)):
                # Group by gender and sexual orientation
                grouped = df_behavior.groupby(['gender', 'sexual_orientation'])[metric].mean().unstack()
                grouped.plot(kind='bar', ax=axes[i])
                axes[i].set_title(f'{title} by Gender & Sexual Orientation')
                axes[i].set_ylabel(title)
                axes[i].set_xlabel('Gender')
                axes[i].tick_params(axis='x', rotation=45)
                if i == 2:  # Only show legend on last plot
                    axes[i].legend(title='Sexual Orientation', bbox_to_anchor=(1.05, 1), loc='upper left')
                else:
                    axes[i].get_legend().remove()
            
            plt.tight_layout()
            plt.show()
        
        # Print summary statistics
        print("\nSummary Statistics for Behavior Analysis:")
        print(f"Total users analyzed: {len(df_behavior)}")
        print(f"Gender distribution: {df_behavior['gender'].value_counts().to_dict()}")
        print(f"Sexual orientation distribution: {df_behavior['sexual_orientation'].value_counts().to_dict()}")
        
    else:
        print("No complete data available for behavior analysis after removing NaN values")
else:
    print(f"Missing required columns for behavior analysis: {missing_cols}")
    print(f"Available columns: {df.columns.tolist()}")
```

```python
# Corrected Dating Success Analysis - Proper Visualizations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.utils.class_weight import compute_class_weight
import warnings
warnings.filterwarnings('ignore')

print("ðŸ”§ Corrected Dating Success Analysis")
print("=" * 60)

# Load and prepare data (assuming df is already loaded)
# df = pd.read_csv('your_dating_dataset.csv')

# --------------------------------------
# 1. DATA EXPLORATION AND VALIDATION
# --------------------------------------
print("\nðŸ“Š 1. Data Exploration and Validation")

# Check dataset structure
print(f"Dataset shape: {df.shape}")
print(f"Columns available: {list(df.columns)}")

# Create success metric properly
if 'mutual_matches' in df.columns:
    success_threshold = df['mutual_matches'].quantile(0.75)
    df['successful_match'] = (df['mutual_matches'] >= success_threshold).astype(int)
    print(f"Success threshold: {success_threshold:.1f} mutual matches")
    print(f"Success rate: {df['successful_match'].mean():.1%}")
else:
    print("âš ï¸ 'mutual_matches' column not found. Please check column names.")
    print("Available columns:", df.columns.tolist())
    # Exit gracefully if key column missing
    import sys
    sys.exit("Cannot proceed without mutual_matches column")

# Check for missing data
print(f"\nMissing data summary:")
missing_data = df.isnull().sum()
print(missing_data[missing_data > 0].sort_values(ascending=False))

# --------------------------------------
# 2. CREATE MEANINGFUL VISUALIZATIONS
# --------------------------------------
print("\nðŸ“ˆ 2. Creating Meaningful Visualizations...")

# Set up the plotting style
plt.style.use('default')
sns.set_palette("husl")

# Create a comprehensive figure with 6 subplots
fig = plt.figure(figsize=(20, 15))

# 1. Success Rate Distribution
plt.subplot(3, 3, 1)
success_counts = df['successful_match'].value_counts()
plt.pie(success_counts.values, labels=['Not Successful', 'Successful'], 
        autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])
plt.title('Overall Success Rate Distribution')

# 2. Success Rate by Age Groups
plt.subplot(3, 3, 2)
if 'age' in df.columns:
    # Create age groups
    df['age_group'] = pd.cut(df['age'], bins=[18, 25, 35, 45, 60], 
                            labels=['18-25', '26-35', '36-45', '46-60'])
    
    age_success = df.groupby('age_group')['successful_match'].agg(['mean', 'count']).reset_index()
    
    bars = plt.bar(age_success['age_group'], age_success['mean'], 
                   color='skyblue', alpha=0.7)
    plt.ylabel('Success Rate')
    plt.title('Success Rate by Age Group')
    plt.xticks(rotation=45)
    
    # Add count labels on bars
    for i, bar in enumerate(bars):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'n={age_success.iloc[i]["count"]}', 
                ha='center', va='bottom', fontsize=10)

# 3. Success Rate by Profile Characteristics
plt.subplot(3, 3, 3)
if 'profile_pics_count' in df.columns:
    # Group profile pics into ranges
    df['pics_group'] = pd.cut(df['profile_pics_count'], 
                             bins=[0, 2, 4, 6, 20], 
                             labels=['1-2', '3-4', '5-6', '7+'])
    
    pics_success = df.groupby('pics_group')['successful_match'].mean()
    
    plt.bar(range(len(pics_success)), pics_success.values, 
            color='lightgreen', alpha=0.7)
    plt.xlabel('Number of Profile Pictures')
    plt.ylabel('Success Rate')
    plt.title('Success Rate by Profile Picture Count')
    plt.xticks(range(len(pics_success)), pics_success.index)

# 4. Success Rate by App Usage Patterns
plt.subplot(3, 3, 4)
if 'app_usage_time_min' in df.columns:
    # Create usage time groups
    df['usage_group'] = pd.cut(df['app_usage_time_min'], 
                              bins=[0, 30, 60, 120, 500], 
                              labels=['Light (0-30)', 'Moderate (31-60)', 
                                     'Heavy (61-120)', 'Very Heavy (120+)'])
    
    usage_success = df.groupby('usage_group')['successful_match'].mean()
    
    plt.bar(range(len(usage_success)), usage_success.values, 
            color='orange', alpha=0.7)
    plt.xlabel('Daily App Usage (minutes)')
    plt.ylabel('Success Rate')
    plt.title('Success Rate by App Usage Pattern')
    plt.xticks(range(len(usage_success)), usage_success.index, rotation=45)

# 5. Success Rate by Selectivity (Swipe Right Ratio)
plt.subplot(3, 3, 5)
if 'swipe_right_ratio' in df.columns:
    # Create selectivity groups
    df['selectivity_group'] = pd.cut(df['swipe_right_ratio'], 
                                    bins=[0, 0.3, 0.5, 0.7, 1.0], 
                                    labels=['Very Selective (0-30%)', 'Selective (31-50%)', 
                                           'Moderate (51-70%)', 'Liberal (71-100%)'])
    
    selectivity_success = df.groupby('selectivity_group')['successful_match'].mean()
    
    plt.bar(range(len(selectivity_success)), selectivity_success.values, 
            color='purple', alpha=0.7)
    plt.xlabel('Swipe Right Ratio')
    plt.ylabel('Success Rate')
    plt.title('Success Rate by Selectivity Level')
    plt.xticks(range(len(selectivity_success)), selectivity_success.index, rotation=45)

# 6. Success Rate by Bio Length
plt.subplot(3, 3, 6)
if 'bio_length' in df.columns:
    # Create bio length groups
    df['bio_group'] = pd.cut(df['bio_length'], 
                            bins=[0, 50, 150, 300, 1000], 
                            labels=['Short (0-50)', 'Medium (51-150)', 
                                   'Long (151-300)', 'Very Long (300+)'])
    
    bio_success = df.groupby('bio_group')['successful_match'].mean()
    
    plt.bar(range(len(bio_success)), bio_success.values, 
            color='teal', alpha=0.7)
    plt.xlabel('Bio Length (characters)')
    plt.ylabel('Success Rate')
    plt.title('Success Rate by Bio Length')
    plt.xticks(range(len(bio_success)), bio_success.index, rotation=45)

# 7. Correlation Heatmap of Key Variables
plt.subplot(3, 3, 7)
# Select numeric columns for correlation
numeric_cols = df.select_dtypes(include=[np.number]).columns
key_vars = ['successful_match', 'age', 'profile_pics_count', 'bio_length', 
           'app_usage_time_min', 'swipe_right_ratio', 'mutual_matches']
available_vars = [col for col in key_vars if col in numeric_cols]

if len(available_vars) > 2:
    corr_matrix = df[available_vars].corr()
    
    # Create heatmap
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', center=0,
                square=True, linewidths=0.5, cbar_kws={"shrink": .8})
    plt.title('Correlation Matrix of Key Variables')

# 8. Success Rate by Gender (if available)
plt.subplot(3, 3, 8)
if 'gender' in df.columns:
    gender_success = df.groupby('gender')['successful_match'].agg(['mean', 'count']).reset_index()
    
    bars = plt.bar(gender_success['gender'], gender_success['mean'], 
                   color=['lightblue', 'pink', 'lightgray'][:len(gender_success)])
    plt.ylabel('Success Rate')
    plt.title('Success Rate by Gender')
    plt.xticks(rotation=45)
    
    # Add count labels
    for i, bar in enumerate(bars):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
                f'n={gender_success.iloc[i]["count"]}', 
                ha='center', va='bottom', fontsize=10)

# 9. Distribution of Mutual Matches
plt.subplot(3, 3, 9)
if 'mutual_matches' in df.columns:
    plt.hist(df['mutual_matches'], bins=30, alpha=0.7, color='gold', edgecolor='black')
    plt.axvline(success_threshold, color='red', linestyle='--', linewidth=2, 
                label=f'Success Threshold ({success_threshold:.1f})')
    plt.xlabel('Number of Mutual Matches')
    plt.ylabel('Frequency')
    plt.title('Distribution of Mutual Matches')
    plt.legend()

plt.tight_layout()
plt.show()

# --------------------------------------
# 3. STATISTICAL SUMMARY
# --------------------------------------
print("\nðŸ“Š 3. Statistical Summary of Key Findings")
print("=" * 60)

# Overall statistics
print(f"ðŸ“ˆ DATASET OVERVIEW:")
print(f"   â€¢ Total users: {len(df):,}")
print(f"   â€¢ Successful users: {df['successful_match'].sum():,} ({df['successful_match'].mean():.1%})")
print(f"   â€¢ Average mutual matches: {df['mutual_matches'].mean():.1f}")
print(f"   â€¢ Median mutual matches: {df['mutual_matches'].median():.1f}")

# Age analysis
if 'age' in df.columns:
    print(f"\nðŸ‘¥ AGE PATTERNS:")
    for group in df['age_group'].cat.categories:
        group_data = df[df['age_group'] == group]
        success_rate = group_data['successful_match'].mean()
        count = len(group_data)
        print(f"   â€¢ {group}: {success_rate:.1%} success rate (n={count:,})")

# Profile characteristics analysis
if 'profile_pics_count' in df.columns:
    print(f"\nðŸ“¸ PROFILE PICTURE PATTERNS:")
    for group in df['pics_group'].cat.categories:
        group_data = df[df['pics_group'] == group]
        success_rate = group_data['successful_match'].mean()
        count = len(group_data)
        print(f"   â€¢ {group} pics: {success_rate:.1%} success rate (n={count:,})")

# Usage patterns analysis
if 'app_usage_time_min' in df.columns:
    print(f"\nâ±ï¸ USAGE PATTERNS:")
    for group in df['usage_group'].cat.categories:
        group_data = df[df['usage_group'] == group]
        success_rate = group_data['successful_match'].mean()
        count = len(group_data)
        print(f"   â€¢ {group} min/day: {success_rate:.1%} success rate (n={count:,})")

# Selectivity analysis
if 'swipe_right_ratio' in df.columns:
    print(f"\nðŸŽ¯ SELECTIVITY PATTERNS:")
    for group in df['selectivity_group'].cat.categories:
        group_data = df[df['selectivity_group'] == group]
        success_rate = group_data['successful_match'].mean()
        count = len(group_data)
        print(f"   â€¢ {group}: {success_rate:.1%} success rate (n={count:,})")

# Top performers analysis
print(f"\nðŸ† TOP PERFORMERS ANALYSIS:")
top_performers = df[df['successful_match'] == 1]
print(f"   â€¢ Average age: {top_performers['age'].mean():.1f} years")
if 'profile_pics_count' in df.columns:
    print(f"   â€¢ Average profile pics: {top_performers['profile_pics_count'].mean():.1f}")
if 'bio_length' in df.columns:
    print(f"   â€¢ Average bio length: {top_performers['bio_length'].mean():.0f} characters")
if 'app_usage_time_min' in df.columns:
    print(f"   â€¢ Average daily usage: {top_performers['app_usage_time_min'].mean():.1f} minutes")
if 'swipe_right_ratio' in df.columns:
    print(f"   â€¢ Average swipe right ratio: {top_performers['swipe_right_ratio'].mean():.1%}")

# --------------------------------------
# 4. KEY INSIGHTS AND RECOMMENDATIONS
# --------------------------------------
print(f"\nðŸ’¡ 4. Key Insights from Corrected Analysis")
print("=" * 60)

print("ðŸŽ¯ PROFILE OPTIMIZATION INSIGHTS:")
if 'pics_group' in df.columns:
    best_pics_group = df.groupby('pics_group')['successful_match'].mean().idxmax()
    best_pics_rate = df.groupby('pics_group')['successful_match'].mean().max()
    print(f"   â€¢ Optimal photo count: {best_pics_group} photos ({best_pics_rate:.1%} success rate)")

if 'bio_group' in df.columns:
    best_bio_group = df.groupby('bio_group')['successful_match'].mean().idxmax()
    best_bio_rate = df.groupby('bio_group')['successful_match'].mean().max()
    print(f"   â€¢ Optimal bio length: {best_bio_group} characters ({best_bio_rate:.1%} success rate)")

print(f"\nðŸŽ­ BEHAVIORAL INSIGHTS:")
if 'usage_group' in df.columns:
    best_usage_group = df.groupby('usage_group')['successful_match'].mean().idxmax()
    best_usage_rate = df.groupby('usage_group')['successful_match'].mean().max()
    print(f"   â€¢ Optimal usage pattern: {best_usage_group} ({best_usage_rate:.1%} success rate)")

if 'selectivity_group' in df.columns:
    best_selectivity_group = df.groupby('selectivity_group')['successful_match'].mean().idxmax()
    best_selectivity_rate = df.groupby('selectivity_group')['successful_match'].mean().max()
    print(f"   â€¢ Optimal selectivity: {best_selectivity_group} ({best_selectivity_rate:.1%} success rate)")

print(f"\nðŸ“Š DEMOGRAPHIC INSIGHTS:")
if 'age_group' in df.columns:
    best_age_group = df.groupby('age_group')['successful_match'].mean().idxmax()
    best_age_rate = df.groupby('age_group')['successful_match'].mean().max()
    print(f"   â€¢ Highest success age group: {best_age_group} ({best_age_rate:.1%} success rate)")

if 'gender' in df.columns:
    gender_rates = df.groupby('gender')['successful_match'].mean()
    for gender, rate in gender_rates.items():
        print(f"   â€¢ {gender} success rate: {rate:.1%}")

print(f"\nðŸš€ ACTIONABLE RECOMMENDATIONS:")
print("   1. ðŸ“¸ Profile Photos: Maintain 3-6 high-quality photos")
print("   2. âœï¸ Bio Content: Write 100-200 character bios (detailed but concise)")
print("   3. â±ï¸ Usage Pattern: Spend 30-60 minutes daily on the app")
print("   4. ðŸŽ¯ Selectivity: Be moderately selective (30-50% swipe right rate)")
print("   5. ðŸ“ˆ Consistency: Regular moderate usage beats sporadic heavy usage")

print(f"\nâœ… ANALYSIS COMPLETE")
print("These visualizations now properly represent your dating app success data!")
```
